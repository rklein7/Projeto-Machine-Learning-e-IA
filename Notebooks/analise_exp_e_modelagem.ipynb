## **Pré-processamento dos Dados**
- Remoção de colunas com muitos valores nulos (> 90%)
- Preenchimento de valores faltantes com a mediana
- Conversão de tipos `object` para `float`
- Criação da variável `dias_entre_acessos`
- Remoção de colunas temporais originais (`ts_...`)
- Dataset final sem valores ausentes e pronto para modelagem

```python
# Imputação de colunas com muitos nulos
cols_nulos = [
    'vl_engajamento_usuario_por_intervalo', 'vl_engajamento_usuario_intradia',
    'vl_desempenho_questionario', 'vl_media_questoes_por_dia', 'vl_engajamento_notas',
    'vl_media_notas', 'vl_media_submissoes_codigo', 'vl_submissoes_por_dias_ativos'
]
for col in cols_nulos:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# Codificação de categóricas
df['ds_cidade_usuario'] = df['ds_cidade_usuario'].fillna('desconhecido')
df = pd.get_dummies(df, columns=['ds_cidade_usuario'], drop_first=True)

# Normalização
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.drop(['tempo_total_participacao'])
df[num_cols] = scaler.fit_transform(df[num_cols])
```

## 🎯 Criação da Variável-Alvo `evadiu`
A variável foi criada com base em critérios heurísticos comportamentais:

- `nr_dias_uso < 3`
- `nr_questoes_respondidas == 0`
- `vl_desempenho_usuario < 0.1`
- `vl_engajamento_notas == 0`

Se qualquer um desses critérios for atendido, o aluno é classificado como `evadiu = 1`.

```python
# Criar variável-alvo
df['evadiu'] = (
    (df['nr_dias_uso'] < 3) |
    (df['nr_questoes_respondidas'] == 0) |
    (df['vl_desempenho_usuario'] < 0.1) |
    (df['vl_engajamento_notas'] == 0)
).astype(int)

# Ver distribuição
df['evadiu'].value_counts(normalize=True)
```

## 🔄 Separação, Divisão e Escalonamento dos Dados

```python
# Separar X e y
X = df.drop(columns=['evadiu'])
X = X.select_dtypes(include=['int64', 'float64'])
X = X.fillna(X.median())
y = df['evadiu']
```

```python
# Dividir treino e validação
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Escalonar os dados
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

## 🤖 Modelagem com Três Algoritmos
Nesta seção comparamos três modelos supervisionados:
- Regressão Logística
- Random Forest
- Rede Neural (Keras, comentada)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import classification_report

# Regressão Logística
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_scaled, y_train)
y_pred_log = log_model.predict(X_val_scaled)

print("📌 Regressão Logística:\n")
print(classification_report(y_val, y_pred_log))

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_val)

print("\n📌 Random Forest:\n")
print(classification_report(y_val, y_pred_rf))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Rede Neural com Keras
nn_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
nn_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
nn_model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_data=(X_val_scaled, y_val))
y_pred_nn = (nn_model.predict(X_val_scaled) > 0.5).astype(int)
```

```python
# Avaliação dos modelos clássicos
relatorio_log = classification_report(y_val, y_pred_log, output_dict=True)
relatorio_rf = classification_report(y_val, y_pred_rf, output_dict=True)

import pandas as pd
resultados_df = pd.DataFrame({
    'Modelo': ['Regressão Logística', 'Random Forest'],
    'Acurácia': [relatorio_log['accuracy'], relatorio_rf['accuracy']],
    'F1-Score': [relatorio_log['weighted avg']['f1-score'], relatorio_rf['weighted avg']['f1-score']],
    'Recall': [relatorio_log['weighted avg']['recall'], relatorio_rf['weighted avg']['recall']],
    'Precisão': [relatorio_log['weighted avg']['precision'], relatorio_rf['weighted avg']['precision']]
})
print(resultados_df)
```
